```python
# Broadcasting (Revenue)
import numpy as np
products = np.array(["Rice", "Oil", "Sugar", "Milk", "Soap"])
quantity = np.array([50, 30, 40, 25, 60])
unit_price = np.array([60, 120, 45, 50, 35])
profit_percent = np.array([10, 15, 12, 8, 20])

# Total Sales = Quantity * Unit Price
total_sales = quantity * unit_price

print("Total Sales:", total_sales)
```

```python
# Array Operations
import numpy as np

total_revenue = np.sum(total_sales)
average_sales = np.mean(total_sales)

print("\nTotal Revenue:", total_revenue)
print("Average Sales:", average_sales)
```

```python
 #Indexing and Slicing

print("\nFirst Product Sales:", total_sales[0])
print("Top 3 Product Sales:", total_sales[:3])
```

```python
#Statistical Operations
print("\nMean Sales:", np.mean(total_sales))
print("Median Sales:", np.median(total_sales))
print("Variance:", np.var(total_sales))
print("Standard Deviation:", np.std(total_sales))
```

```python
#Profit Calculation
import numpy as np
profit_percent = np.array([10, 15, 12, 8, 20])

# Profit using vectorized computation
profit = total_sales * profit_percent / 100

print("\nProfit for each product:", profit)
```

```python
# Interpretation
import numpy as np
products = np.array(["Rice", "Oil", "Sugar", "Milk", "Soap"])
max_sale_index = np.argmax(total_sales)

print("\nBest Selling Product:", products[max_sale_index])
print("Highest Sales Value:", total_sales[max_sale_index])
```
